# Nasa Mars News

```python
#dependencies
import os
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
#URL to be scraped
url ='https://redplanetscience.com/'
browser.visit(url)
```

```python
#retrieve page with the results module
html = browser.html
```

```python
#create beautiful soup object; parse with html.parser
soup = bs(html, 'html.parser')
```

```python
#print bs object
print(soup.prettify())
```

```python
results = soup.find('div', class_="content_title")
print(results)
```

```python
#store thread title
news_title = results.text
       
print(news_title)
```

```python
#title element is class = "card-text" 
p_results = soup.find('div', class_="article_teaser_body")
print(p_results)
```

```python
#store thread title
news_p = p_results.text
       
print(news_p)
```

# JPL Mars Space Images - Featured Image

```python
#URL to be scraped
space_url = 'https://spaceimages-mars.com/'
browser.visit(space_url)
```

```python
#print bs object
print(soup.prettify())
```
